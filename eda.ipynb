```python
data_temp_zuz = pd.read_csv('temp_zuz.csv', sep=';')
data_temp_zuz['Czas'] = pd.to_datetime(data_temp_zuz['Czas'])
data_temp_zuz['delta_czas'] = data_temp_zuz['Czas'].diff()
data_temp_zuz['delta_temp_zuz'] = data_temp_zuz['temp_zuz'].diff()

# Filter out troublesome rows
data_temp_zuz = data_temp_zuz[data_temp_zuz['Czas'].dt.strftime('%M:%S').eq('00:00')]
data_temp_zuz = data_temp_zuz[data_temp_zuz['delta_czas'].dt.days == 0]
data_temp_zuz = data_temp_zuz[data_temp_zuz['delta_czas'].dt.seconds == 3600]

data_temp_zuz = data_temp_zuz.set_index('Czas')
```
